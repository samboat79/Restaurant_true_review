```python
#Get list of Kansas City Restaurants, their latitude, longitude and cuisine type

pages = [0, 20, 40, 60, 80]

name = []
lat = []
long = []
cuisines = []
coordinates = []

for page in pages:
    url = "https://developers.zomato.com/api/v2.1/search?"
    city_search = "entity_id=856&entity_type=city&start="
    rest_url = f"{url}&apikey={api_key}&{city_search}{page}"
    rest_response = requests.get(rest_url).json()
    i = 0
    for x in rest_response['restaurants']:
        name.append(rest_response['restaurants'][i]['restaurant']['name'])
        lat.append(rest_response['restaurants'][i]['restaurant']['location']['latitude'])
        long.append(rest_response['restaurants'][i]['restaurant']['location']['longitude'])
        cuisines.append(rest_response['restaurants'][i]['restaurant']['cuisines'])

        i+=1

coordinates = list(zip(lat,long))   
print(coordinates)
```

```python
# create dataframe
str(coordinates).strip('"')

import pandas as pd
restaurant_df = pd.DataFrame({
    "Restaurant_Name": name,
    "Latitude": lat,
    "Longitude": long,
    "Cuisine": cuisines,
    "Coordinates": coordinates
})

```

```python
restaurant_df.to_csv('restaurant.csv')
```
