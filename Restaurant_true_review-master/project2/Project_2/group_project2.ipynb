```python
#import and initialize Sentiment Initializer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
import json
import pandas as pd
```

```python
#how do i reference the restaurant.Json file
#Do i have to choose a sample string from the file? 
with open('tweets.JSON', 'r') as file:
   json_data = json.load(file)
```

```python
#create three empty list
compound = []
pos = []
neu = []
neg = []
restaurants = []


```

```python
for TweetData in json_data:
    print(TweetData)
    results = analyzer.polarity_scores(TweetData["Tweets"])
    pos.append(results["pos"])
    compound.append(results["compound"])
    neu.append(results["neu"])
    neg.append(results["neg"])
    restaurants.append(TweetData["Restaurants"])
    #print(results)
```

```python
tweets_one_df = pd.DataFrame({
    "positive": pos,
    "negative": neg,
    "compound": compound,
    "neutral" : neu,
    "Restaurants" : restaurants
})
```
